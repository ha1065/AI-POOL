```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from statistics import mean
```

```python
def best_fit_slope(X, y):
    #print(X, y)
    m = (((mean(X)*mean(y)) - mean(X*y))/
        ((mean(X)*mean(y)) - mean(X*X)))
    return m
```

```python
def best_fit_slope_intercept(X, y):
    m = (((mean(X)*mean(y)) - mean(X*y))/
        ((mean(X)*mean(y)) - mean(X*X)))
    
    b = mean(y) - m * mean(X)
    return m, b
```

```python
def calc_epoch_slope(data):
    slopes = []
    for row in range(len(data)+1):
        #print(row)
        if row > 1:
            m = best_fit_slope(data[:row]["Epoch"], data[:row]["Losses"])
            slopes.append([row+1, m])
    return pd.DataFrame(slopes, columns = ["Epoch", "Slope"])
        
    
```

```python
#MSE LOSS: L200
#Slopes: S200
def plot_loss_slopes(L, S):
    
    fig, ax1 = plt.subplots()
    
    
    color = 'tab:blue'
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('MSE Loss', color=color)
    ax1.plot(L["Epoch"], L["Losses"], color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    
    ax2 = ax1.twinx()
    
    color = 'tab:red'
    
    ax2.set_ylabel('LR Slope', color=color)
    ax2.plot(S["Epoch"], S["Slope"], color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    
#     fig.tight_layout()
    plt.show()
    
    
    
```

```python
def add_epochs(L, epochs):
    index = []
    for i in range(1,epochs+1):
        index.append(i)
    L["Epoch"] = np.asarray(index)
    return L
```

```python
def plot_loss_RL(L, x, y):
    
    fig, ax1 = plt.subplots()
    
    
    color = 'tab:blue'
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('MSE Loss', color=color)
    ax1.plot(L["Epoch"], L["Losses"], color=color)
    ax1.tick_params(axis='y', labelcolor=color)
    
    color = 'tab:orange'
    
    ax1.plot(x, y, color=color)
    
#     fig.tight_layout()
    plt.show()
```
